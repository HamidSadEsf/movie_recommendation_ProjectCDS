```python
# List of movie IDs present in df
movie_ids_in_df = df.index.tolist()

# Filter movies in df_rating that are present in df
df_rating_filtered = df_rating[df_rating['movieId'].isin(movie_ids_in_df)]

# Group by userId and count the number of ratings
user_ratings_count = df_rating_filtered.groupby('userId')['rating'].count()

# Filter users based on the ratings count condition
filtered_users = user_ratings_count[(user_ratings_count > 150) & (user_ratings_count < 200)].index

# Filter df_rating to keep only the rows for filtered users
df_rating_final = df_rating_filtered[df_rating_filtered['userId'].isin(filtered_users)]

# Group by userId and aggregate the movieIds as a list
user_rated_movies = df_rating_final.groupby('userId')['movieId'].apply(list)

# Create the final DataFrame
result_df = pd.DataFrame({'RatedMovies': user_rated_movies})

result_df
```

```python
#import shap
import numpy as np
from functools import partial
from model.ContentBasedRec import recommendation
from data_script.Preprocess_Content_Based import get_df

# Assuming df_movies is your DataFrame with movie vectors
# and df_result is your DataFrame with user rated movies
df = get_df()

# Define a partial function with fixed arguments
#partial_recommendation = partial(recommendation, number_of_recommendations=0, num_clusters=18)

# Initialize the SHAP explainer with a background dataset of zeros
#explainer = shap.Explainer(partial_recommendation, masker=shap.maskers.Independent(np.zeros_like(df.values)))

# Create an array to accumulate SHAP values for each tag
tag_accumulated_shap = np.zeros(df.shape[1])

# Iterate through each user's rated movies
for row in result_df.RatedMovies:
    # Calculate the original recommendation scores
    original_scores = recommendation(row, 0, 18, lambda_val=0.5).score
    
    # Iterate through each tag for perturbation analysis
    for tag_index, tagId in enumerate(df.columns):
        # Perturb the tag relevance scores
        mean_vector = df.loc[row].mean().copy()
        perturbed_mean_vector[tagId] = mean_vector[tagId] + 0.5
        
        # Convert perturbed_mean_vector to an array
        #perturbed_mean_vector_array = perturbed_mean_vector.values.reshape(1, -1)
        
        # Calculate the SHAP values for the perturbed input
        #shap_values = explainer(perturbed_mean_vector_array, max_evals=2 * len(df.columns) + 1)  # Increase max_evals
        
        # Calculate the contribution of the tag as the difference in recommendation scores
        tag_contribution = original_scores - recommendation(perturbed_mean_vector, 0, 18, lambda_val=0.5).score
        tag_accumulated_shap[tag_index] += tag_contribution

# Calculate average contribution for each tag
average_tag_contributions = tag_accumulated_shap / len(df_result)

# Normalize and visualize average tag contributions
normalized_contributions = (average_tag_contributions - average_tag_contributions.min()) / (average_tag_contributions.max() - average_tag_contributions.min())

# Print or visualize the normalized average tag contributions
print(normalized_contributions)
```
